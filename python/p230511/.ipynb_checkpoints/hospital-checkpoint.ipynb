# 1. 병원정보 서비스 읽어 오기
    - 변수 저장

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('병원정보서비스 2021.12.csv', encoding='cp949')
hospital = pd.DataFrame(data)
hospital
```

```python
hospital.columns
```

```python
hospital.drop(columns=['암호화요양기호','종별코드','시도코드','시군구코드','x좌표','y좌표'], inplace=True)
```

```python
hospital.info()
```

```python
hospital.describe()
```

```python
#지역별 상급종합병원의 수 데이터 -> 시각화
```

```python
g_t= hospital[hospital.종별코드명 == '상급종합']['시도코드명'].value_counts()
```

```python
g_t=g_t.sort_index().sort_values(ascending=False)
```

```python
plt.rcParams['font.family'] = 'D2coding'
plt.rcParams['axes.unicode_minus'] = False
plt.rc('font',size=10)
plt.rc('axes',titlesize=12)

plt.tight_layout(pad=2) 
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

plt.subplot(1,2,1)
g_t.T.plot(
    kind = "pie",
    subplots=True,
    ax=axes[0],
    ylabel='',
    legend=False,
    colormap = 'Greens',
    autopct='%.f%%',
    title=['지역별 상급종합병원의 수'],
    shadow=True,
)

plt.subplot(1, 2, 2)
g_t.T.plot(
    kind='bar',
    subplots=True, 
    colormap = 'cool',
    ax=axes[1],
    xlabel= '',
    title =[''],
    
)

plt.legend(['상급종합'])
plt.xticks(rotation=0)
plt.savefig('지역별 상급병원.jpg')
```
