## seaborn 사용한 시각화
    - matplotlib를 기반으로 한 파이썬 시각화 라이브러리
    - matplotlib의 확장판
    - 간결한 문법
    - DataFrame의 데이터를 바로 집계하여 쉽게 시각화 : 그룹화 과정 생략 가능
---
    - 오차막대
    - 신뢰구간을 의미
    - 0~100의 값을 가짐
    - 기본값 95
    - 즉 추론은 100번 중에 95번 정도 맞다
    - 막대 그래프 위에 표시
    - 부트 스트랩 신뢰구간(Bootstrap confidence interval)
    - 이 데이터를 기반으로 유사한 상황의 95%가 이 범위 내에서 결과를 얻음
    - 신뢰구간은 mean, median(중앙값), mode(최빈값) emd rhfugkdu rPtks
    - 설정 가능한 값 : ci, pi, se, sd, None
---
- estimator
    - np.mean
    - np.max
    - np.min
    - np.median
    - np.sum
    - np.count_nonozen

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks")

# Create a dataset with many short random walks
rs = np.random.RandomState(4)
pos = rs.randint(-1, 2, (20, 5)).cumsum(axis=1)
pos -= pos[:, 0, np.newaxis]
step = np.tile(range(5), 20)
walk = np.repeat(range(20), 5)
df = pd.DataFrame(np.c_[pos.flat, step, walk],
                  columns=["position", "step", "walk"])

# Initialize a grid of plots with an Axes for each walk
grid = sns.FacetGrid(df, col="walk", hue="walk", palette="tab20c",
                     col_wrap=4, height=1.5)

# Draw a horizontal line to show the starting point
grid.refline(y=0, linestyle=":")

# Draw a line plot to show the trajectory of each random walk
grid.map(plt.plot, "step", "position", marker="o")

# Adjust the tick positions and labels
grid.set(xticks=np.arange(5), yticks=[-3, 3],
         xlim=(-.5, 4.5), ylim=(-3.5, 3.5))

# Adjust the arrangement of the plots
grid.fig.tight_layout(w_pad=1)
```

#### 가설2 코로나 기간(2020~ 2022)에 서울, 경기 평당 분양가 상승률이 낮았을 것이다?
- 서울은 연도별 평당분양가 2018년부터 2019년까지 가장 높았고 점차 상승률 낮아 지고, 2020부터 2021년까지 11% 하락이 되었다. 그러나 2021년부터 2022년 까지는 2020년도에 비해 약 15%가 올랐다. 
- 경기는 연도별 평당분양가가 2017년부터 2018년 약 16% 상승이었다가 점차 줄어드는 성장세를 보이다. 2019년부터 2021년까지 각년도 3% 하락을 한다. 그러나 2021년부터 소폭 상승세로 되어 상승세가 2%로가 늘었다.
