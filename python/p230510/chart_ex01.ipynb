```python
# 원형 그래프
data = [10,20,30,5]
labels=['하나','둘','셋','넷']
#값이을 표현을 labels로 해용
_, _, pie =plt.pie(data, 
        labels=labels, 
        autopct='%.1f%%', 
        startangle=45, #첫번째 조각의 시작 각도(위치)
#         counterclock=False, # 시계 반향으로 했어요
        explode=[0.23,0,0,0], # 원하는 조각을 다른 조각과 다르게 하여, 강조 형태
        shadow=True, #그림자
        colors=['r','pink','gold','silver'],#각 도형 색깔
        wedgeprops={'width': 0.6, 'edgecolor':'silver','linewidth':2}, 
                   #각 파트별 모양 {} 튜플형 width : 반지름에서 0.6
                   #'edgecolor':'silver' 테두리 컬러
                   #'linewidth':2 조각 테두리 
       )
#값의 비율 표시해줘용 autopct = '%f'실수 '%d' 정수 .1 소수 처음자리 까지
#startangle = 45 => 45도로 시작해요
#
for t in pie: 
    t.set_color('w')
# 
plt.show()
```

```python
# 원형 그래프
# 코드 정리

data = [10,20,30,5]
labels=['하나','둘','셋','넷']
explode=[0.23,0,0,0] # 원하는 조각을 다른 조각과 다르게 하여, 강조 형태
colors=['r','pink','gold','silver']#각 도형 색깔
wedgeprops={'width': 0.6, 'edgecolor':'silver','linewidth':2}
    #각 파트별 모양 {} 튜플형 width : 반지름에서 0.6
    #'edgecolor':'silver' 테두리 컬러
    #'linewidth':2 조각 테두리 

#값 표현을 labels로 해용

_, _, pie =plt.pie(data,
        labels=labels,
        autopct='%.1f%%', 
        startangle=45,
#         counterclock=False, # 시계 반향으로 했어요
        explode=explode, 
        shadow=True, #그림자
        colors=colors,
        wedgeprops=wedgeprops
       )
#값의 비율 표시해줘용 autopct = '%f'실수 '%d' 정수 .1 소수 처음자리 까지
#startangle = 45 => 45도로 시작해요
#
for t in pie: 
    t.set_color('w')
# 
plt.show()
```

## 데이터 분포를 시각적으로 볼 수 있는 bins 값이다

- 즉, bins 매개변수는 히스토그램의 막대 개수를 조절하는 역할을 합니다. 막대의 개수가 많아질수록 데이터의 분포를 더 세밀하게 파악할 수 있지만, 반대로 막대의 개수가 적을수록 전체적인 분포의 특성을 파악할 수 있습니다. bins 값의 선택은 데이터의 특성과 분석 목적에 따라 조절할 수 있습니다.

```python
## 히스토그램
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)
# bins는 구간을 5개로 나눌꺼에용
plt.hist(data, bins=10, label='bins=10')

plt.hist(data, bins=30, label='bins=30')

plt.legend()
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

## 데이터 1000가지
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) +1

#히스토그램

plt.figure() #차트 설정 
plt.hist((data1,data2),bins=33, label=('data1','data2'), histtype='bar')
plt.title('histtype- bar')
plt.legend()

plt.figure() #차트 새로 만들기
plt.hist([data1,data2],bins=33, label=('data1','data2'), histtype='barstacked')
#barstacked는 누적되는 형태
plt.title('histtype- barstacked')
plt.legend()

plt.figure() 
plt.hist([data1,data2],bins=33, label=('data1','data2'), histtype='step')
plt.title('histtype- step')
#점으로만 그리기
plt.legend()

plt.figure()
plt.hist([data1,data2],bins=33, label=('data1','data2'), histtype='stepfilled')
plt.title('histtype- stepfilled')
#점으로 그린 차트에 색 채워 넣기


plt.legend() #항상 figure로 생성전에 넣어줘야함
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

## x 축
x = np.arange(0,100)

## y 축
y1 = [v*v for v in x]
y2 = [-v*v for v in x]
y3 = [v*v/2 for v in x]



#그래프 그리기
fig, (ax1,ax2) = plt.subplots(2,1)
fig.subplots_adjust(hspace=0.3)
fig.set_size_inches(10, 8)


#첫번쨰 subplot
ax1.plot(x,y1,x,y3)
ax1.set_xlabel('x축')
ax1.set_ylabel('y축')
ax1.grid()

#두번째 subplot
ax2.plot(x,y2)
ax2.set_title('제목')


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

## x 축
x = np.arange(0,100)

## y 축
y1 = [v*v for v in x]
y2 = [-v*v for v in x]
y3 = [v*v/2 for v in x]



#그래프 그리기
fig, axs = plt.subplots(2,1)
fig.subplots_adjust(hspace=0.3)
fig.set_size_inches(10, 8)


#첫번쨰 subplot
axs[0].plot(x,y1,x,y3)
axs[0].set_xlabel('x축')
axs[0].set_ylabel('y축')
axs[0].grid()

#두번째 subplot
axs[1].plot(x,y2)
axs[1].set_title('제목')


plt.show()
```
