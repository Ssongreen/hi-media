```python
import pandas as pd
import matplotlib.pyplot as plt

# csv 읽어오기

titanic = pd.read_csv('titanic.csv')
titanic[:2]

# 1pclass에 생존 비율 ? 다른 pcalss 보다 높다?

pclass_p = titanic.groupby('Pclass').Survived.value_counts(normalize=True).sort_index() * 100 

plt.rcParams['font.family'] = 'malgun gothic'
plt.rcParams['axes.unicode_minus'] = False

fig, ax = plt.subplots(1, 3)

ax[0].pie(pclass_p[1],
          labels=['사망자', '생존자'],
          autopct='%.0f%%',
          shadow=True,
          startangle=90,
          colors=['Tomato', 'DodgerBlue'],
          wedgeprops={'width': 0.6, 'linewidth': 2}
          )
ax[0].set_title('1등실')

ax[1].pie(pclass_p[2],
          labels=['사망자', '생존자'],
          autopct='%.0f%%',
          shadow=True,
          startangle=90,
          colors=['Tomato', 'DodgerBlue'],
          wedgeprops={'width': 0.6, 'linewidth': 2}
          )
ax[1].set_title('2등실')

ax[2].pie(pclass_p[3],
          labels=['사망자', '생존자'],
          autopct='%.0f%%',
          shadow=True,
          startangle=90,
          colors=['Tomato', 'DodgerBlue'],
          wedgeprops={'width': 0.6, 'linewidth': 2}
          )
ax[2].set_title('3등실')

# plt.show()
plt.savefig('티켓별 생존자 비율.jpg')
```

- 가설2 티켓 별로 생존자가 다르며, 1등석이 가장 높을 것이다.
    - 생존자 342명 중 1등석 216명중 136명(63%) 생존, 2등석 184명중 87명 생존(47%), 3등석 487명 중 119명(24%)으로 1등석이 가장 많은 생존률이었고 티켓별로 생존률이 높았다.
