```python
#문제1 20세 이하인 선수 정보 추출
market[market['age']<= 20]
```

```python
#문제2 20세 이하인 선수 숫자만
market[market['age'] <= 20].age.count()
```

```python
#문제3 22세인 선수의 이름, 나이, 국가 추출
market[market['age'] == 22][['name','age','nation']]
```

```python
#각 나이 인원수
market.age.value_counts().sort_index()
```

```python
#문제4 축구 선수의 평균 나이
import math
mean = market['age'].mean()
sum(market.age) / len(market.age)
```

```python
#문제5 축구 선수의 최고 나이 
market['age'].max()
```

```python
#문제6 축구 선수의 최소 나이
market['age'].min()
```

```python
#문제7 평균 나이 이상인 선수의 이름, 나이, 국가
market[market['age'] >= mean][['name','age','nation']]
market[market.loc[:,'age'] >= mean][['name','age','nation']]
```

```python
#문제8 한국 선수 정보 추출
market[market['nation'] == str('Korea, South')]
market[market['nation'].str.contains('Kor')]
market[market['nation'].apply(lambda x: 'kor' in x.lower())]
market[market['nation'].str.match('.*kor', case=False)]
```

```python
#문제9 토트넘 소속인 선수 정보 추출
market[market['club'] == str('Tottenham Hotspur')]
market[market['club'].str.contains('tot', case=False)]
market[market['club'].apply(lambda x:'tot' in x.lower())]
market[market['club'].str.match('.*tot', case=False)]
```

```python
#이적료의 평균
market.value.mean()
```

```python
#24세 평균 이적료
market[market.age == 24].value.mean()
# 소수 첫째자리, 반올림 round()
market[market.age == 24].value.mean().round()
# 소수 둘째자리, 반올림 round(1)
market[market.age == 24].value.mean().round(1)
# 정수 반올림 round(-1)
market[market.age == 24].value.mean().round(-1)
```

```python
#그룹화는 오름차순으로 묶는다.
#<pandas.core.groupby.generic.DataFrameGroupBy object at 0x0000012601711520> 기준점이 많아서 그래용
groub_age = market.groupby(market.age)
groub_age
```

```python
for idx, group in groub_age:
    print('그룹화 기준: 나이', idx)
    print(group)
    print('='*45)
```
