```python
"""
1 - 10page(200개) 뉴스 타이틀을 화면에 출력하기.
- 출력 형태 :1. 타이틀 

"""
import json
import requests
from bs4 import BeautifulSoup 
```

```python
headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36"}

json_url = "https://news.naver.com/main/mainNews.naver?sid1=101&date=%2000:00:00:00&page="

for i in range(1,11):
    alls = requests.get(f"{json_url}{i}", headers = headers)
    
    news =alls.json()
    news_page = json.loads(news["airsResult"])
    
    for idx , result in enumerate(news_page["result"]["101"],1):
        print(f"{idx}.{result['title']}")
```
