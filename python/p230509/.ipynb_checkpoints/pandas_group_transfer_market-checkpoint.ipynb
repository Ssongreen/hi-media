```python
import pandas as pd
```

```python
market = pd.read_csv('market100.csv')
market[:4]
```

```python
group_age = market.groupby('age')
group_age
```

```python
#그룹으로 묶은 오름차순으로 보여진다. 
# 기준으로 묶은 값으로 DataFrame이다.
# 그룹안에 기준으로 묶은 값도 데이터가 들어가 있다.
for group in group_age:
    print(group)
```

```python
# get_group(18) key 형태로 해당이 되는 데이터를 볼 수 있음. items랑 같다.
group_age.get_group(18)
```

```python
#나이별 선수수
group_age.count().number
```

```python
#내림차순
group_age.number.count().sort_values(ascending=False)
```

```python
group_age.size()
```

```python
#국가별 선수
market_nation = market.groupby('nation')
market_nation.size()
```

```python
# 국가별 선수의 이적료 합
market_nation.sum('value')['value']
market_nation.value.sum().sort_values(ascending=False)
```

```python
# 국가별 선수의 이적료 평균
#평균 값은 소속된 인원수에 따라 다르기 떄문에 높다라고 볼 수 없어용
market_nation.mean(numeric_only=True)['value'].round(2).sort_values(ascending=False)
```
