# 행 또는 열 추가

```python
import pandas as pd

## 데이터프레임 생성

## 수식 함수 

df = pd.DataFrame(
    [
        range(1,21,2),
        range(2,21,2),
        range(10,101,10)
    ],
    
    columns=[f'co{i}' for i in range(1,11)]
)
df
```

## 행추가
- 데이터 마지막 쪽에 행이 추가해요

```python
df.loc[3] = [r for r in range(100, 110, 1)]
```

```python
df
```

```python
df.loc[6] = [r for r in range(200,210,1)]
```

```python
df
```

### iloc로는 row 추가가 되지 않아용 , 덮어 쓰기가 되용

```python
df.iloc[5] = [r for r in range(300,310,1)]
```

```python
df.iloc[3] = [r for r in range(400,410,1)]
```

```python
df
```

```python
#컬럼수가 맞아야 되용
# df.loc[100] = [500, 5001 ]
# df
```

### 빈행 추가

```python
df
```

### 결측치가 들어 오면 실수로 포함이 되기에 결측치가 오면 모든 값을 실수로 나타낸다

```python
df.loc[12] = 7
df
```

```python
df.loc[9] = pd.Series()
# nan 은 결측치이다. null 과 같이 없는 값이다.
```

```python
df.info()
#인덱스와 non-null로 뺴주면 결측치가 있는 개수가 나와용
```

## 열추가
- 원본 추가

### 마지막 위치에 추가

```python
df['co11'] = [1,2,4,6,7]
df
```

```python
# 변수명으로 사용은 앙되용
# df.co12 = [1,5,7,84,2]
```

```python
# 전체적으로 5로 추가가 됩니다.
df['co12'] = 5
```

```python
df
```

### 특정 위치에 추가
- 인덱스로 좌표를 넣는다.

```python
#allow_duplicates 는 중복 허용
df.insert(1, 'newjeans', [6,7,84,-2,31])
```
