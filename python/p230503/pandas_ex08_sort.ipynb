```python
import pandas as pd

## 데이터프레임 생성

## 수식 함수 

df = pd.DataFrame(
    [
        range(1,21,2),
        range(2,21,2),
        range(10,101,10)
    ],
    
    columns=[f'co{i}' for i in range(1,11)]
)
df
```

## 정렬 culmn 기준으로 오름 차순

```python

df.sort_values('co2')
```

```python
help(df.sort_values)
```

## 정렬 culmn 기준으로 내림 차순

```python
df.sort_values('co2', ascending=False)
```

## 행 기준 오름 차순

- 행 오름 차순

```python
df.sort_values(1,axis=1)
```

## 행 기준 내림 차순

```python
# index 1의 값을 내림차순

df.sort_values([1],axis=1, ascending=False)
```

## 인덱스 정렬 

### 날짜로 정렬로 인덱스 내림 정렬을 주로 해용

### 행기준

#### 오름 차순

```python
df.sort_index()
```

#### 내림 차순

```python
df.sort_index(ascending=False)
```

### 열기준

#### 내림차순

```python
# 정렬은 자리수로 비교해서 그래용
# 열 이름의 첫 글자부터 ASCII 문자 코드에 따라 정렬

df.sort_index(axis=1) 
```

#### object 가 str과 비슷하다

#### 내림차순

```python
df.sort_index(axis=1, ascending=False)
```

```python
help(df.sort_index)
```

#### 문제 열의 인덱스를 다음과 같이 정렬해용
- co10 ~ co1

```python
df[df.columns[::-1]].sort_index(key=lambda x: x.astype(str), ascending=False, axis=1)
```
