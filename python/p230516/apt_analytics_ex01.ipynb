```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
apt_price =pd.read_csv('../p230515/apt_price.csv')
apt_price
```

```python
plt.rcParams['font.family'] = 'D2coding'
plt.rcParams['axes.unicode_minus'] = False
sns.violinplot(
    apt_price, 
    x='연도',
    y='분양가격',
).figure.set_size_inches(12,6)
plt.grid(axis='y', alpha=0.5)
```

```python
sns.boxplot(
    apt_price,
    x='연도',
    y='분양가격'
).figure.set_size_inches(12,6)
plt.grid()
#결측치를 0로 넣는다는것은 최솟값이 나타나기에 좋지 않게 보인다.
```

```python
sns.kdeplot(
     apt_price,
    x='연도',
    y='분양가격',
).figure.set_size_inches(12,6)
plt.grid()
```

```python
#문제 
sns.lineplot(
    apt_price,
    x='연도',
    y='분양가격',
    hue='지역명',
    errorbar=None,
).figure.set_size_inches(12,6)
plt.grid()
plt.legend(loc=(1,0))
```

```python
#문제 
sns.pointplot(
    apt_price,
    x='연도',
    y='분양가격',
    hue='지역명',
    errorbar=None,
).figure.set_size_inches(12,6)
plt.grid()
plt.legend(loc='upper right',
          bbox_to_anchor=(1.2,1),
          )
```

```python
g_local_year_price_mean= apt_price.groupby(['지역명','연도']).분양가격.mean().astype(int).unstack()
g_local_year_price_mean
```

```python
g_local_year_price_mean.index[:]
```

```python
x_labels = g_local_year_price_mean.index
```

```python
g_local_year_price_mean.plot(
    figsize=(12,6),
    grid=1,
    xticks= range(len(x_labels))
)
```
