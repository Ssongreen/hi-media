```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.DataFrame({
    'n1' : [1,5,10,15,20],
    'n2' : [1,2,3,4,20], #값이 갑짜기 튈때, 이상값, 극단값이라고 불림
    'n3' : [1,17,18,19,20]
    
})
df
```

## 이상치 
- 비정상적인 극단값, 극단적인 값
- 이상치로 평균값이 상승 또는하락된 현상을 편향(skew,왜도)되었다고 함
- 편향된 데이터와 이상치로 정보가 왜곡되면, 전형적인 값 파악을 위해 중앙 값 등 사용
---
- 이상치를 변경해서 바꿀지를 고민하라.

```python
# 평균 값과 중앙값하고 차이난다면, 이상치가 있다고 볼 수 있다.
# 평균이 중앙값이 높다면, 큰값이 있고
# 평균이 중앙값보다 낮으면, 낮은 값이 있다.
df.describe().round(1)
```

# kdeplot

```python
#KDE
# 커널 밀도 추정
# Kernel Density Estimation
sns.kdeplot(
    df,
)

plt.grid()
```

```python
sns.kdeplot([1,2,3,4,5])
```

```python
df.n1.hist(bins=3)
```

```python
df.describe().round(1)
```

```python
# 오른쪽 편향 왜도 skewness
#positive = skewness 
# 이상치들의 꼬리가 오른쪽으로 길게나타남
# 평균 값을 오른쪽을 잡아끔
# 평균값이 대부분의 값보다 높음
sns.kdeplot(df.n2,
           
           )
plt.grid()
plt.axvline(x=df.n2.mean(),
            color='g',
           linewidth=2,
            linestyle='--'
           )
#중앙값
plt.axvline(x=df.n2.median(),
           color='b',
            linewidth=2,
            linestyle='--'
           )
```

```python
#왼쪽 편향
sns.kdeplot(df.n3,
           
           )
plt.grid()

plt.axvline(x=df.n3.mean(),
           color='g',
            linewidth=2,
            linestyle='--'
           )
plt.axvline(x=df.n3.median(),
            color='b',
            linewidth=2,
            linestyle='--'
           )
```

```python
sns.kdeplot(df.n1
)
plt.grid()

plt.axvline(x= df.n1.mean(),
           color='g',
            linewidth=2,
            linestyle='--',
           )

plt.axvline(x=df.n1.median(),
            color='b',
            linewidth=2,
            linestyle='--'
           )
```

```python
sns.kdeplot(data=df,y='n1',
           fill=True, #속 채우기
           color='r',
           )
```

```python
sns.displot(data=df.n1)
```
