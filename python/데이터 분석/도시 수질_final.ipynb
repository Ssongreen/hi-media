```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
#파일 읽어 오기
city =pd.read_csv('도시수질19~22(전처리).csv',index_col=0)
```

```python
city.columns
```

```python
plt.rcParams['font.family'] = 'D2coding'
plt.rcParams['axes.unicode_minus'] = False

winter_columns = [col for col in city.columns if col.endswith(('.01', '.02', '.12'))]
summer_columns = [col for col in city.columns if col.endswith(('.06', '.07', '.08'))]

c_w = city[['항목'] + winter_columns]
c_s = city[['항목'] + summer_columns[:-1]] 

# 평균값
winter_data = c_w[c_w.항목.isin(['수온'])].mean(numeric_only=True)
summer_data = c_s[c_s.항목.isin(['수온'])].mean(numeric_only=True)

plt.rcParams['font.family'] = 'D2coding'
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(10, 6))

x = np.arange(len(winter_data.index))

width = 0.35

bar_winter = plt.bar(x, winter_data, width, label='동계(온도)')
bar_summer = plt.bar(x + width, summer_data, width, label='하계(온도)')

x_labels = [f'{winter_col} / {summer_col}' for winter_col, summer_col in zip(winter_data.index, summer_data.index)]

for rect in bar_winter:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size=12)

for rect in bar_summer:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size=12)

plt.xlabel('날짜 (동계 / 하계)')
plt.ylabel('평균')
plt.title('동계와 하계 온도')
plt.legend(loc=(1, 0.1))
plt.xticks(x + width / 2, x_labels, rotation=45)
plt.show()
```
