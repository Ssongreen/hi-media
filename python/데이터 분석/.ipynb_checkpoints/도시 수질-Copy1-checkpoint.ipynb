```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
#파일 읽어 오기
city =pd.read_csv('도시수질19~22(전처리).csv',index_col=0)
```

```python
plt.rcParams['font.family'] = 'D2coding'
plt.rcParams['axes.unicode_minus'] = False

# 하계, 동계, 컬럼 추출
winter_columns = [col for col in city.columns if col.endswith(('.01', '.02', '.12'))]
summer_columns = [col for col in city.columns if col.endswith(('.06', '.07', '.08'))]

c_w = city[['항목'] + winter_columns]
c_s = city[['항목'] + summer_columns]

# 평균값
winter_data = c_w[c_w.항목.isin(['수온'])].mean(numeric_only=True)
summer_data = c_s[c_w.항목.isin(['수온'])].mean(numeric_only=True)

plt.figure(figsize=(10, 6))

x = np.arange(len(winter_data.index))

width = 0.35

bar_winter = plt.bar(x, winter_data, width, label='동계(온도)')
bar_summer = plt.bar(x + width, summer_data, width, label='하계(온도)')

x_labels = [f'{winter_col} / {summer_col}' for winter_col, summer_col in zip(winter_data.index, summer_data.index)]

for rect in bar_winter:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size=12)

for rect in bar_summer:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size=12)

plt.xlabel('날짜 (동계 / 하계)')
plt.ylabel('평균')
plt.title('동계와 하계 온도')
plt.legend(loc=(1, 0.1))
plt.xticks(x + width / 2, x_labels, rotation=45)
plt.show()
```

```python
import folium

rivers = {
    '공촌천': (37.743437, 126.849032),
    '남천': (35.821701, 128.657473),
    '무거천': (36.755869, 126.366379),
    '수영강': (35.155144, 129.114972),
    '유등천': (36.115319, 128.354046),
    '전주천': (35.822662, 127.144657)
}

center = [36.5, 127.5]

map = folium.Map(location=center, zoom_start=8)

for location, coord in rivers.items():
    lat, lon = coord
    folium.Marker(location=[lat, lon], popup=location).add_to(map)
map

```

```python
from mpl_toolkits.basemap import Basemap

# 지도 생성
map = Basemap(llcrnrlon=124, llcrnrlat=33, urcrnrlon=131, urcrnrlat=39, resolution='i')

# 지도 요소 추가
map.drawcoastlines(color='green')
map.drawcountries(color='yellow')
map.fillcontinents(color='lightgray', zorder=0)

# 좌표 데이터를 반복하며 마커 생성
for location, coord in rivers.items():
    lat, lon = coord
    x, y = map(lon, lat)
    plt.plot(x, y, 'bo', markersize=8, label=location)

# 범례 추가
plt.legend(loc='lower left')

# 그래프 출력
plt.show()
```
